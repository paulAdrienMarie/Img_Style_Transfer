### Test with 512 x 512 images

```python
import matplotlib.pyplot as plt
import numpy as np
from torchvision import transforms

import sys
from pathlib import Path

src_path = Path("../src")  # Chemin relatif vers le rpertoire `src`
sys.path.append(str(src_path))

from style_transfer import style_transfer


def imshow(tensor, title=None):
    image = tensor.cpu().clone()
    image = image.squeeze(0)
    image = transforms.ToPILImage()(image)
    plt.imshow(image)
    if title is not None:
        plt.title(title)
    plt.axis("off")

# Combinaisons de poids de style et contenu
style_weights = [10 ** 2, 10 ** 4, 10 ** 6, 10 ** 8]
content_weight = 1

content_image_name = "content.jpeg"
style_image_name = "style.jpeg"

plt.figure(figsize=(15, 10))

for i, style_weight in enumerate(style_weights):
    print(f"Processing style weight: {style_weight}")
    output = style_transfer(content_image_name, style_image_name, style_weight, content_weight)
    
    plt.subplot(2, 2, i + 1)
    imshow(output, title=f"Style Weight: {style_weight:.0e}")

plt.tight_layout()
plt.show()
```
